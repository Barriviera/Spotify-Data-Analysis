```python
import os
import pandas as pd
```

```python
diretorio = "./DataFrames/Musicas"
arquivos = [arquivo for arquivo in os.listdir(diretorio) if not arquivo.startswith('.')] #Ignora arquivos escondidos

df_musicas = pd.DataFrame()

for arquivo in arquivos:
    data_atual = pd.read_csv(diretorio+"/"+arquivo)
    df_musicas = pd.concat([df_musicas, data_atual])
```

```python
df_toptracks = df_musicas.sort_values('popularity',ascending = False).head(100)

df_toptracks_view = df_toptracks[['track_name', 'popularity']].copy()

df_toptracks_view.rename(columns={"track_name":"Tracks","popularity":"Popularity"},inplace=True)

df_toptracks_view.reset_index()[['Tracks', 'Popularity']]
```

```python
# Assuming your dataframe is named 'df' with columns 'artist' and 'popularity'
df_grouped_artist = df_musicas.groupby('artist_name')['popularity'].sum().reset_index()

# Sort the dataframe based on the sum of popularity in descending order
df_artist_sorted = df_grouped_artist.sort_values(by='popularity', ascending=False)

# Display the most popular artists
most_popular_artists = df_artist_sorted['artist_name'].reset_index()['artist_name']
most_popular_artists.head(100)
```

```python
# Assuming your dataframe is named 'df' with columns 'artist' and 'popularity'
df_grouped_genres = df_musicas.groupby('artist_genres')['popularity'].sum().reset_index()

# Sort the dataframe based on the sum of popularity in descending order
df_genres_sorted = df_grouped_genres.sort_values(by='popularity', ascending=False)

# Display the most popular artists
most_popular_genres = df_genres_sorted['artist_genres'].reset_index()['artist_genres']
most_popular_genres.head(100)
```

```python
df_longestTracks = df_musicas[["track_name","duration"]].sort_values(by="duration",ascending=False)[:10]

df_longestTracks.reset_index()[['track_name', 'duration']]
```
